# Demo jax giving a gradient of a function at any point

Here we have a function `someFn` which is discontinous -- it contains an `if` statement. Jax is able to evaluate gradient for both code paths, automatically


```python
def someFn(x):
    if x>1.0:        
        return -3.0*x+x*x
    else:
        return(jax.numpy.sqrt(x))
    
y_result=jax.numpy.array([jax.value_and_grad(someFn)(x) for x in xbase])
px_show(px.line(x=xbase,y=[y_result[:,0], y_result[:,1]]))
```

demo some functional constructs:

```python
value_y, gradient_y = map(numpy.array, zip(*map(jax.value_and_grad(someFn), xbase)))
```

Yeah, functional transforms are "powerful" in sense that they are excellent in making the code unreadable.
